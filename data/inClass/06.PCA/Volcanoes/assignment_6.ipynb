```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Load the Data

```python
df = pd.read_csv('alaskaPCA.csv')
df.head()
```

## Visualizations

```python
plt.figure(figsize=(8, 5))
plt.hist(df['zmean_m'], bins=40, edgecolor='black')
plt.xlabel('Mean Elevation (m)')
plt.ylabel('Frequency')
plt.title('Distribution of Mean Elevation')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.scatter(df['area_km2'], df['slope_deg'], alpha=0.5, s=5)
plt.xlabel('Area (kmÂ²)')
plt.ylabel('Slope (degrees)')
plt.title('Glacier Area vs Slope')
plt.show()
```

## Visualization

```python
plt.figure(figsize=(8, 5))
plt.hist(df['aspect_deg'], bins=40, edgecolor='black')
plt.xlabel('Aspect (degrees)')
plt.ylabel('Frequency')
plt.title('Distribution of Glacier Aspect')
plt.show()
```
